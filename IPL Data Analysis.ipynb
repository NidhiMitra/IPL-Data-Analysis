```python
# extracxting the records where a team won first batting
batting_first = ipl[ipl['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
#making a histogram
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs'])
plt.xlabel("Runs")
plt.title("DISTRIBUTION OF RUNS")
plt.show()
```

```python
#after batting first ,the no of wins w.r.t batting_first
batting_first['winner'].value_counts()
```

```python
 #making bar plot fot the top 3 teams after batting first
plt.figure(figsize=(7,5))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=["red","blue","yellow"])
plt.show()
```

```python
#making a pie chart
plt.figure(figsize=(6,6))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
